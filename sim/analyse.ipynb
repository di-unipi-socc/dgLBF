## Import libraries and read the data

Data are divided into 2 dataframes. 

- `all-norel.parquet`: results for the *original* methodoology (_without_ reliability requirements),
- `all.parquet`: results using the *new* methodology and reliability requirements, in terms of _path protection_ and _anti-affinity_.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

FIG_SIZE = (15, 6)

df = pd.read_parquet('results/all.parquet')
df_norel = pd.read_parquet('results/all-norel.parquet')

TOPOLOGIES = df["Topology"].unique()

print(df.shape, df_norel.shape)
```

```python
df_success = df[~df['Output'].isin(['no_result', 'timeout'])]
df_fails = df[df['Output'].isin(['no_result', 'timeout'])]

df_success_norel = df_norel[~df_norel['Output'].isin(['no_result', 'timeout'])]
df_fails_norel = df_norel[df_norel['Output'].isin(['no_result', 'timeout'])]

# remove all the lines where Nodes is in [16, 32]
TO_REMOVE = [16, 32]
df_success = df_success[~df_success['Nodes'].isin(TO_REMOVE)]
df_fails = df_fails[~df_fails['Nodes'].isin(TO_REMOVE)]
df_success_norel = df_success_norel[~df_success_norel['Nodes'].isin(TO_REMOVE)]
df_fails_norel = df_fails_norel[~df_fails_norel['Nodes'].isin(TO_REMOVE)]
```

```python
from typing import Callable

def by_topology(df_all: pd.DataFrame, fn: Callable, **kwargs):
    for topology in TOPOLOGIES:
        for rp in list(df_all["RepProb"].unique()):
            df_topo = df_all[(df_all['Topology'] == topology) & (df_all['RepProb'] == rp)]
            fn(df_topo, title=f"Topology {topology}, RepProb {rp}", **kwargs)

    fn(df_all, title='All Topologies', **kwargs)
```

```python
def percentile(df_raw: pd.DataFrame, field: str, percentile: float = 0.90):
    
    df_raw.loc[:, "percentile"] = df_raw.groupby("Nodes")[field].transform(lambda x: x.quantile(percentile))
    df_raw = df_raw[df_raw[field] <= df_raw["percentile"]]

    return df_raw
```

```python
from scipy.stats import zscore

def remove_outliers_zscore(df_raw: pd.DataFrame, field: str , threshold: int = 3):
    z = np.abs(zscore(df_raw[field]))
    df_raw.loc[z > threshold, field] = df_raw[field].median()
    return df_raw
```

# EXECUTION TIMES

#### BY TOPOLOGY

```python
def execution_times(df_et: pd.DataFrame, title: str):
    plt.figure(figsize=FIG_SIZE)
    sns.lineplot(
        data=df_et,
        x="Flows",
        y="Time",
        hue="Nodes",
        style="Nodes",
        markers=True,
        markersize=9,
        errorbar=None,
        dashes=False,
        palette="colorblind",
    )

    # Set plot title and labels
    plt.title(title)
    plt.xlabel('Flows')
    plt.ylabel('Time')
    plt.legend(title='Nodes')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"plots/{title}.pdf", format='pdf', dpi=600)
    plt.show()

by_topology(df_success, execution_times)
```

#### CATPLOT

```python
# execution_times catplot, directly in the code cell not with a function
execution_times_catplot = sns.catplot(
    data=df_success,
    x="Flows",
    y="Time",
    hue="Topology",
    row="RepProb",
    col="Nodes",
    kind="point",
    palette="colorblind",
    legend_out=True,
    height=6,
    aspect=1.5,
    hue_order=sorted(TOPOLOGIES),
)

for ax in execution_times_catplot.axes.flat:
    ax.grid(True)
    ax.set_xscale("log")
    # ticks = list(sorted(df_success["Flows"].unique()))
    # show_ticks = [i for i in ticks if i % 500 == 0]

    # ax.set_xticks(list(range(len(ticks))))
    # ax.set_xticklabels([str(t) if t in show_ticks else "" for t in ticks])
# Set plot title and labels
plt.title('Execution Times')
plt.xlabel('Flows')
plt.ylabel('Time')
plt.grid(True)
plt.tight_layout()
plt.savefig(f"plots/execution_times.pdf", format='pdf', dpi=600)
plt.show()
```

# TIME PER FLOW

#### BY TOPOLOGY

```python
# Compute the time spent for 1 flow
df1 = df_success.copy()
df1['Time_per_Flow'] = df1['Time'] / df1['Flows']

def time_per_flow(df_tpf: pd.DataFrame, title: str):
    plt.figure(figsize=FIG_SIZE)
    sns.lineplot(
        data=df_tpf,
        x="Flows",
        y="Time_per_Flow",
        hue="Nodes",
        style="Nodes",
        markers=True,
        markersize=9,
        errorbar=None,
        dashes=False,
        palette="colorblind"
    )

    # Set plot title and labels
    plt.title(title)
    plt.xlabel('Flows')
    plt.ylabel('Time per Flow')
    plt.legend(title='Nodes')
    plt.grid(True)
    plt.show()


by_topology(df1, time_per_flow)
```

#### CATPLOT 

```python
# Compute the time spent for 1 flow
df1 = df_success.copy()
df1['Time_per_Flow'] = df1['Time'] / df1['Flows']

time_per_flow_catplot = sns.catplot(
    data=df1,
    x="Flows",
    y="Time_per_Flow",
    hue="Topology",
    row="RepProb",
    col="Nodes",
    kind="point",
    palette="colorblind",
    legend_out=True,
    height=6,
    aspect=1.5,
    hue_order=sorted(TOPOLOGIES),
)

for ax in time_per_flow_catplot.axes.flat:
    ax.grid(True)
    # ax.set_xscale("log")

plt.title('Time per Flow')
plt.xlabel('Flows')
plt.ylabel('Time per Flow')
plt.grid(True)
plt.tight_layout()
plt.savefig(f"plots/time_per_flow.pdf", format='pdf', dpi=600)
plt.show()
```
